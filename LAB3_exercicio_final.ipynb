```python
# Show a plot of the response rate by the sales channel.

data1["sales channel"].unique()
```

```python
# Created a variable for each channel from the column 'Sales Channel'. 
# In this variable I filtered from my Dataframe for each channel.
# Calculated the 'yes' % of each channel and divided it by the total and multiplied by 100.
#Created 2 final variables to being able to make the plot bar

agent = data1[data1['sales channel'].isin(['Agent'])]
call_Center = data1[data1['sales channel'].isin(['Call Center'])]
branch = data1[data1['sales channel'].isin(['Branch'])]
web = data1[data1['sales channel'].isin(['Web'])]

agent_Resp_Rate = (agent['response'].value_counts()['Yes'])/agent['response'].count()*100
call_Center_Resp_Rate = (call_Center['response'].value_counts()['Yes'])/call_Center['response'].count()*100
branch_Resp_Rate = (branch['response'].value_counts()['Yes'])/branch['response'].count()*100
web_Resp_Rate = (web['response'].value_counts()['Yes'])/web['response'].count()*100

sales1_channels=['Agent', 'Call_Center', 'Branch', 'Web' ]
rate = [agent_Resp_Rate,call_Center_Resp_Rate,branch_Resp_Rate,web_Resp_Rate] 
```

```python
#Plot Bar creation

plt.bar(x=sales1_channels, height= rate )
plt.xlabel("Sales Channel")
plt.ylabel("Response Rate [%]")
plt.ylim(0, 20)
plt.show() 
```

```python
# Show a plot of the response rate by the total claim amount.

#Created bins for the colum Total claim amout

bins = [0, 500, 1000, 1500, 2000, 2500, 3000]

data1['total claim amount bins'] = pd.cut(data1['total claim amount'], bins)

#Created a groupby variable to performe an agrupation of values from 'response' and the 'total claim amount bins', and then all those values were multiplied by 100
#Created a column 'percentage_yes' to represent the above countage
#Filtered all the results in which the response was yes

percent_yes = data1['response'].groupby(data1['total claim amount bins']).value_counts(normalize=True)*100
percent_yes = percent_yes.reset_index(name='percentage_yes')
percent_yes = percent_yes[percent_yes['response'] == 'Yes']

#Plot Bar creation

plt.bar(percent_yes['total claim amount bins'].astype(str), percent_yes['percentage_yes'])
plt.xlabel("Total Claim Amount")
plt.ylabel("Response rate")
plt.title("Response rate by Total Claim Amount")
plt.show()
```

```python
# Show a plot of the response rate by income.

# The algorithm present in this exercise was the same as the last one.

bins = [0, 25000, 50000, 100000]

data1['income bins'] = pd.cut(data1['income'], bins)

percent_yes2 = data1['response'].groupby(data1['income bins']).value_counts(normalize=True)*100
percent_yes2 = percent_yes2.reset_index(name='percentage_yes')
percent_yes2 = percent_yes2[percent_yes2['response'] == 'Yes']

##Plot Bar creation

plt.bar(percent_yes2['income bins'].astype(str), percent_yes['percentage_yes'])
plt.xlabel("Income")
plt.ylabel("Response rate")
plt.title("Response rate by Income")
plt.show()
```
