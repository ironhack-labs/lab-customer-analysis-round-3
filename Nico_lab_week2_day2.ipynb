```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Load the dataset
file1 = pd.read_csv("marketing_customer_analysis.csv")
file1
```

```python
# Show DataFrame info
file1.info()
```

```python
#Describe DataFrame
file1.describe()
```

```python
#Show a plot of the total number of responses
sns.countplot(x='Response', data=file1)
plt.title('Total Number of Responses')
plt.show()
```

```python
#Show a plot of the response rate by the sales channel
sns.countplot(x='Sales Channel', hue='Response', data=file1)
plt.title('Response Rate by Sales Channel')
plt.show()
```

```python
#Show a plot of the response rate by the total claim amount
sns.histplot(x='Total Claim Amount', hue='Response', data=file1, element='step', stat='probability', common_norm=False)
plt.title('Response Rate by Total Claim Amount')
plt.show()
```

```python
#Show a plot of the response rate by income
sns.histplot(x='Income', hue='Response', data=file1, element='step', stat='probability', common_norm=False)
plt.title('Response Rate by Income')
plt.show()
```
