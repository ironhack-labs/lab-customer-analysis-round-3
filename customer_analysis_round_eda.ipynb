This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing first data cleaning, formatting, and structuring.

Start by reading the dataset, standardizing column names, dealing with null values and creating the month column. You can do so by using the functions you developed in previous labs, or by reading the clean dataset you saved in the last lab `data structuring and combining`.

```python
import pandas as pd 
import numpy as np

marketing_analysis = pd.read_csv('marketing_customer_analysis.csv')
display(marketing_analysis)
```

Then, answer the following questions:

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
yes_1000_customers = marketing_analysis[(marketing_analysis['total_claim_amount'] > 1000) & (marketing_analysis['response'] == 'Yes')]
display(yes_1000_customers)
display(yes_1000_customers)
display(yes_1000_customers.describe().T)
display(yes_1000_customers.describe(include='object'))
```

2. Analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign.

```python
average_claim_amount = yes_1000_customers.groupby(['policy_type', 'gender'])['total_claim_amount'].mean()
display(round(average_claim_amount))
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
print()
customer_count_by_state = marketing_analysis.groupby('state').size()
print('Count of customers by state is:')
display(customer_count_by_state)
print()
filtered_states = customer_count_by_state[customer_count_by_state > 500]
print('The states with more than 500 customers are:')
display(filtered_states)
more_than_500_states = marketing_analysis[marketing_analysis['state'].isin(filtered_states.index)]
display(more_than_500_states)
display(round(more_than_500_states.describe().T))
display(more_than_500_states.describe(include='object'))
```

4. Find the median customer lifetime value by education level and gender.

```python
median_lifetime_value = marketing_analysis.groupby(['education', 'gender'])['customer_lifetime_value'].median()
display(median_lifetime_value)
```

5. Find the maximum, minimum, and average number of open complaints by state and policy type.

```python
complaint_values = marketing_analysis.groupby(['state', 'policy_type'])['number_of_open_complaints'].sum()
complaint_values_stats = marketing_analysis.groupby(['state', 'policy_type'])['number_of_open_complaints'].agg(['max', 'min', 'mean'])
print('The number of Open Complaints by state and policy type is:')
display(complaint_values)
display(complaint_values_stats)
```

6. The marketing team wants to analyze the number of policies sold by state and month.

Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
marketing_analysis['effective_to_date'] = pd.to_datetime(marketing_analysis['effective_to_date'])
marketing_analysis['month'] = marketing_analysis['effective_to_date'].dt.month
state_month_groups = marketing_analysis.groupby(['state', 'month']).size().reset_index(name='number_of_policies_sold')
policies_by_state_month = state_month_groups.pivot(index='state', columns='month', values='number_of_policies_sold')
policies_by_state_month = policies_by_state_month.dropna()
display(policies_by_state_month)
```

7. Display a new DataFrame that contains the number of policies sold by month by state for the top 3 states with the highest number of policies sold.

*Hint:* 
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
marketing_analysis['effective_to_date'] = pd.to_datetime(marketing_analysis['effective_to_date'])
marketing_analysis['month'] = marketing_analysis['effective_to_date'].dt.month
state_month_grouped = marketing_analysis.groupby(['state', 'month']).size().reset_index(name='number_of_policies_sold')
sorted_sales= state_month_grouped.sort_values('number_of_policies_sold', ascending=False)
top_3_sales_by_states = sorted_sales['state'].unique()[:3]
top_3_sales_by_states_filtered = sorted_sales[sorted_sales['state'].isin(top_3_sales_by_states)]
final_dataframe = top_3_sales_by_states_filtered.pivot(index='state', columns='month', values='number_of_policies_sold')
final_dataframe = final_dataframe.dropna()
print(final_dataframe)
```

8. The marketing team wants to analyze the effect of different marketing channels on the customer response rate.

Hint: You can use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") by marketing channel.

External Resources for Data Filtering: https://towardsdatascience.com/filtering-data-frames-in-pandas-b570b1f834b9

```python
analyze = ['response','sales_channel']
marketing_analysis_selected = marketing_analysis[analyze]
marketing_analysis_filtered = marketing_analysis_selected[marketing_analysis_selected['response'] == 'Yes']
response_rate = marketing_analysis_filtered.groupby('sales_channel').size() / marketing_analysis.groupby('sales_channel').size()
response_rate = response_rate.reset_index(name='customer_response_rate')

display(response_rate)
```

Show a plot of the total number of responses.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fig, ax = plt.subplots()
ax.bar(x = marketing_analysis['response'].value_counts().index, height = marketing_analysis['response'].value_counts().values, color = ['maroon', 'olivedrab'])
ax.set_xlabel("Response")
ax.set_ylabel("Count")
ax.set_title("Count of Responses")
plt.show()
```

Show a plot of the response rate by the sales channel.

```python
response_rate = marketing_analysis.groupby('sales_channel')['response'].value_counts(normalize=True).mul(100).rename('response_rate').reset_index()
positive_responses = response_rate[response_rate['response'] == 'Yes']
plt.figure(figsize=(10, 6))
sns.barplot(data=positive_responses, x='sales_channel', y='response_rate')
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

Show a plot of the response rate by the total claim amount.

```python
num_bins = 10
bin_edges = np.linspace(marketing_analysis['total_claim_amount'].min(), marketing_analysis['total_claim_amount'].max(), num_bins+1)
labels = [f"{bin_edges[i]:.2f}-{bin_edges[i+1]:.2f}" for i in range(num_bins)]
marketing_analysis['claim_amount_range'] = pd.cut(marketing_analysis['total_claim_amount'], bins=bin_edges, labels=labels, include_lowest=True)
response_rate = marketing_analysis.groupby('claim_amount_range')['response'].value_counts(normalize=True).mul(100).rename('response_rate').reset_index()
positive_responses = response_rate[response_rate['response'] == 'Yes']
plt.figure(figsize=(10, 6))
plt.bar(positive_responses['claim_amount_range'], positive_responses['response_rate'])
plt.title('Response Rate by Total Claim Amount')
plt.xlabel('Total Claim Amount Range')
plt.ylabel('Response Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

Show a plot of the response rate by income.

```python
income_ranges = [0, 35000, 50000, 65000, 80000, 95000, 110000]
income_labels = ['0-35k', '35k-50k', '50k-65k', '65k-80k', '80k-95k', '95k-110k']
marketing_analysis['income_range'] = pd.cut(marketing_analysis['income'], bins=income_ranges, labels=income_labels, include_lowest=True)
response_rate = marketing_analysis.groupby('income_range')['response'].value_counts(normalize=True).mul(100).rename('response_rate').reset_index()
positive_responses = response_rate[response_rate['response'] == 'Yes']
plt.figure(figsize=(10, 6))
plt.bar(positive_responses['income_range'], positive_responses['response_rate'])
plt.title('Response Rate by Income')
plt.xlabel('Income Range')
plt.ylabel('Response Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
