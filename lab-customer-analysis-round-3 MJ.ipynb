```python
import pandas as pd


Mark_data_A3 = pd.read_excel("marketing_customer_analysis.xlsx")

Mark_data_A3
```

```python
# Proper the column

cols = []

for i in range(len(Mark_data_A3.columns)) :
    cols.append(Mark_data_A3.columns[i].lower().replace(' ','_'))
Mark_data_A3.columns = cols

Mark_data_A3
```

```python
# replace name for column employmentstatus

Mark_data_A3= Mark_data_A3.rename(columns={'employmentstatus':'employment_status'} )

Mark_data_A3

```

```python
# importing the needed librairie


import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

```

```python
# plot to see the Data for response, Histograme looks the best for binary Data

sns.histplot(Mark_data_A3['response'])
plt.show()
```

```python
# adding a column with 0 and 1 to count response and make a plot

Mark_data_A3['response_nb'] = np.where(Mark_data_A3['response'] == 'Yes',1 , 0)



Mark_data_A3
```

```python
# check unique value in column sales channel

Mark_data_A3['sales_channel'].unique()
```

```python
# Group the data by 'sales_channel' and count the 'response_nb' values for 'Yes'
grouped = Mark_data_A3.groupby(['sales_channel', 'response_nb'])['response_nb'].count()

# Access the count of 'Yes' responses (1) for each sales channel
response_counts_web = grouped.get(('Web', 1), 0)
response_counts_agent = grouped.get(('Agent', 1), 0)
response_counts_branch = grouped.get(('Branch', 1), 0)
response_counts_Call_ctr = grouped.get(('Call Center', 1), 0)


# count total of row for each sales channel

Total_agent = Mark_data_A3.groupby('sales_channel').size()['Agent' ]

Total_branch = Mark_data_A3.groupby('sales_channel').size()['Branch']

Total_call_center = Mark_data_A3.groupby('sales_channel').size()['Call Center']

Total_web = Mark_data_A3.groupby('sales_channel').size()['Web']

# transform in pourcent by dividing response yes count, by total row for each sales channel, multiplied by 100 to get pourcent

Prct_web_Y = (response_counts_web / Total_web ) * 100
Prct_agent_Y = (response_counts_agent/ Total_agent) * 100
Prct_branch_Y =(response_counts_branch / Total_branch ) * 100
prct_Call_ctr_Y = (response_counts_Call_ctr / Total_call_center) * 100

# creating a new Data Frame with new value obtain previously, x each sale channel, y response rate

data = {
    'Sales_Channel': ['Web', 'Agent', 'Branch', 'Call Center'],
    'Response_Rate_(%)': [Prct_web_Y, Prct_agent_Y, Prct_branch_Y, prct_Call_ctr_Y]
}

Response_rate_df = pd.DataFrame(data)

sns.barplot(x=Response_rate_df['Sales_Channel'], y=Response_rate_df['Response_Rate_(%)'], data=Response_rate_df, estimator='mean') 

plt.xlabel('Sales Channel')
plt.ylabel('Response Rate (%)')
plt.title('Response Rate by Sales channel')
plt.show()




```

```python
# change the data type for column total_claim_amount, to remove decimal to have integrer , to better groupby after


Mark_data_A3['total_claim_amount'] = Mark_data_A3['total_claim_amount'].astype(int)

Mark_data_A3
```

```python
# check data types

Mark_data_A3.dtypes
```

```python

# Groupe by total calim amount and count number of response,using .count, the result is a serie
grouped1 = Mark_data_A3.groupby(['response','total_claim_amount'])['response_nb'].count()

# creating a new serie from the result of previous line with pandas
series = pd.Series(grouped1)

# reseting the index of the serie to make anew data frame
Df_response_rate = series.reset_index()
                               
# filter by yes response the dataframe created by the result 
yes_responses = Df_response_rate[Df_response_rate['response'] == 'Yes']
# creating a new column response rate with column response nb and sum of all the response_nb 
yes_responses['response_rate'] = (yes_responses['response_nb'] / Df_response_rate['response_nb'].sum()) * 100

# rounding pourcentage result to seconmd decimal
yes_responses['response_rate'] = yes_responses['response_rate'].round(2)


# creating a bar plot as it seems the best way to look at data, tried squatter plot but was not really readable 
sns.barplot(x=yes_responses['response_rate'], y=yes_responses['total_claim_amount'] ) 

# settings the names of axis, and graph 
plt.xlim(-1,6)
plt.ylim(0 , 600)
plt.xlabel('Response Rate (%)')
plt.ylabel('Total Claim Amount in $')
plt.title('Response Rate by Total Claim Amount')
plt.show()
print(Df_response_rate)
print(yes_responses)
```

```python
# Groupe by income and count number of response,using .count, the result is a serie
grouped2 = Mark_data_A3.groupby(['response','income'])['response_nb'].count()

# creating a new serie from the result of previous line with pandas
series = pd.Series(grouped2)

# reseting the index of the serie to make a new data frame for income by response rate
Df_response_income = series.reset_index()

                               
# filter by yes response the dataframe created by the result 
yes_responses = Df_response_income[Df_response_income['response'] == 'Yes']
# creating a new column response rate with column response nb and sum of all the response_nb 
yes_responses['response_rate'] = (yes_responses['response_nb'] / Df_response_income['response_nb'].sum()) * 100

# rounding pourcentage result to seconmd decimal
yes_responses['response_rate'] = yes_responses['response_rate'].round(3)


# creating a sactter plot 
sns.scatterplot(x=yes_responses['income'], y=yes_responses['response_rate'] ) 

# settings the names of axis, and graph 
plt.xlabel('Income in $')
plt.ylabel('Response Rate (%)')
plt.title('Response Rate by Income')
plt.show()
print(Df_response_income)
print(yes_responses)
```
