```python
# 3.Show a plot of the total number of responses.

response_counts = data['response'].value_counts()
response_counts.plot(kind = 'bar')
plt.title('Number of Responses') 
plt.xlabel('Number')
plt.ylabel('Yes/No')
plt.show()
```

```python
# 4. Show a plot of the response rate by the sales channel. 

sales_channel = []
response_rate = []
for channel in data['sales channel'].unique():
    sales_channel.append(channel)
    filtered=data[(data['sales channel']==channel)]
    total_response = filtered['response'].count()
    yes_response=filtered['response'].value_counts()['Yes']
    percentage = yes_response/total_response*100
    response_rate.append(percentage)
```

```python
sales_channel
```

```python
response_rate
```

```python
plt.bar(x=sales_channel, height=response_rate)
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate [%]')
plt.show
```

```python
# 5. Show a plot of the response rate by the total claim amount.

bins = [0, 500, 1000, 1500, 2000, 2500, 3000]

data['total claim amount bins'] = pd.cut(data['total claim amount'], bins)

percent_yes = data['response'].groupby(data['total claim amount bins']).value_counts(normalize=True).mul(100).reset_index(name='percentage_yes')
percent_yes = percent_yes[percent_yes['response'] == 'Yes']
```

```python
plt.bar(percent_yes['total claim amount bins'].astype(str), percent_yes['percentage_yes'])
plt.xlabel("Total Claim Amount")
plt.ylabel("Response Rate")
plt.title("Response Rate by Total Claim Amount")
plt.show()
```

```python
# 6. Show a plot of the response rate by income.

bins = [0, 25000, 50000, 100000]

data['income bins'] = pd.cut(data['income'], bins)

percent_yes2 = data['response'].groupby(data['income bins']).value_counts(normalize=True).mul(100).reset_index(name='percentage_yes')
percent_yes2 = percent_yes2[percent_yes2['response'] == 'Yes']
```

```python
plt.bar(percent_yes2['income bins'].astype(str), percent_yes['percentage_yes'])
plt.xlabel("Income")
plt.ylabel("Response Rate")
plt.title("Response Rate by Income")
plt.show()
```
