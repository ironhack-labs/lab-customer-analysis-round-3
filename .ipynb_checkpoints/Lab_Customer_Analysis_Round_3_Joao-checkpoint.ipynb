```python
# Import libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import file
df = pd.read_excel(r"C:\Users\joaoa\Desktop\Ironhack\Labs\lab-customer-analysis-round-3\files_for_lab\excel_files\marketing_customer_analysis.xlsx")
df
```

```python
# Standardize columns names
cols = []
for i in range(len(df.columns)):
    cols.append(df.columns[i].lower().replace(" ","_"))
df.columns = cols
df
```

```python
# Looking for null values
print("Total NaN values:", df.isna().sum().sum())
df.isna().sum()
```

```python
# Drop duplicates
df = df.drop_duplicates()
df
```

```python
# Plot of the total number of responses

sns.histplot(df["response"])

plt.xlabel("Type of responses")
plt.ylabel("Number of responses")
plt.title("Total number of responses")
plt.show()
```

```python
# Looking for min values, max values and unique values

print("Min of total claim amount:", min(df["total_claim_amount"]))
print("Max of total claim amount:", max(df["total_claim_amount"]))
print("Min of income:", min(df["income"]))
print("Max of income:", max(df["income"]))
print("Unique values of sales channel:", df["sales_channel"].unique())
```

```python
# Creating a new dataframe for response rate graphs

df2 = pd.DataFrame()

df2["sales_channel"] = df["sales_channel"].unique()
count = 0
for value in df2["sales_channel"].values.tolist():
    df2.loc[count, "sales_channel_yes_(%)"] = 100*len(df[(df["response"] == "Yes") & (df["sales_channel"] == value)]) / len(df[(df["sales_channel"] == value)])
    count += 1
    
n1_bins = int(input('Select the number of bins for the "Response rate by the total claim amount" bar plot: '))
count = 0
for _ in range(0,n1_bins):
    min_interval_1 = count*max(df["total_claim_amount"])/n1_bins
    max_interval_1 = (count+1)*max(df["total_claim_amount"])/n1_bins
    df2.loc[count, "claim_amount"] = "[" + str(int(min_interval_1)) + ";" + str(int(max_interval_1)) + "["
    df2.loc[count, "claim_amount_yes_(%)"] = 100*len(df[(df["response"] == "Yes") & (df["total_claim_amount"] >= min_interval_1) & (df["total_claim_amount"] < max_interval_1)]) / len(df[(df["total_claim_amount"] >= min_interval_1) & (df["total_claim_amount"] < max_interval_1)])    
    count += 1

n2_bins = int(input('Select the number of bins for the "Response rate by income" bar plot: '))
count = 0
for _ in range(0,n2_bins):
    min_interval_2 = count*max(df["income"])/n2_bins
    max_interval_2 = (count+1)*max(df["income"])/n2_bins
    df2.loc[count, "income"] = "[" + str(int(min_interval_2)) + ";" + str(int(max_interval_2)) + "["
    df2.loc[count, "income_yes_(%)"] = 100*len(df[(df["response"] == "Yes") & (df["income"] >= min_interval_2) & (df["income"] < max_interval_2)]) / len(df[(df["income"] >= min_interval_2) & (df["income"] < max_interval_2)])    
    count += 1
    
df2
```

```python
# Plot of the response rate by sales channel

sns.barplot(x = df2["sales_channel"], y = df2["sales_channel_yes_(%)"], data = df2)
plt.xlabel("Sales channel")
plt.ylabel("Response rate (%)")
plt.title("Response rate by the sales channel")
plt.show()
```

```python
# Plot of the response rate by total claim amount

plot1 = sns.barplot(x = df2["claim_amount"], y = df2["claim_amount_yes_(%)"], data = df2)
plt.setp(plot1.get_xticklabels(), rotation=90)
plt.xlabel("Total claim amount")
plt.ylabel("Response rate (%)")
plt.title("Response rate by the total claim amount")
plt.show()
```

```python
# Plot of the response rate by income

plot2 = sns.barplot(x = df2["income"], y = df2["income_yes_(%)"], data = df2)
plt.setp(plot2.get_xticklabels(), rotation=90)
plt.xlabel("Income")
plt.ylabel("Response rate (%)")
plt.title("Response rate by income")
plt.show()
```
