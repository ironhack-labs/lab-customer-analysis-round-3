### Show a plot of the total number of responses.

```python
sns.histplot(data['response'],color='green')
plt.xlabel('Responses')
plt.ylabel('Quantity')
plt.title('Total number of responses')
plt.show()
```

### Show a plot of the response rate by the sales channel.

```python
group = data.groupby('sales_channel')['response'].value_counts(normalize=True).unstack().fillna(0)
group
```

```python
group['response_rate']=group['Yes']*100

group
```

```python
group.reset_index(inplace=True)
```

```python
group
```

```python
sns.barplot(x= group['sales_channel'], y= group['response_rate'], data=group)

plt.show()


```

### Show a plot of the response rate by the total claim amount.

```python
group1 = data.groupby('total_claim_amount')['response'].value_counts(normalize=True).unstack().fillna(0)
group1
```

```python
group1.reset_index(inplace=True)
```

```python
group1['response_rate']=group1['Yes']*100
```

```python
group1
```

```python
group1.describe().T
```

```python
bins = [-1, 500,1000,1500,2000,2500,3000] 
labels = ["[0:500]","]500:1000]","]1000:1500]","]1500:2000]","]2000:2500]", "]2500:3000]",]
group1['total_claim_amount_class'] = pd.cut(group1['total_claim_amount'], bins=bins, labels=labels)
```

```python
group1
```

```python
sns.barplot(x= group1['total_claim_amount_class'], y= group1['response_rate'] , data=group1)
plt.xlabel('Total claims amount')
plt.ylabel('Response rate')
plt.title('Response rate by Total claims amount')
plt.show()


####  One improvement can be remove some outliers since the Q3 is ~550 wich is 1/6 of the 2893.239678
```

### Show a plot of the response rate by income.

```python
group2 = data.groupby('income')['response'].value_counts(normalize=True).unstack().fillna(0)
group2
```

```python
group2.reset_index(inplace=True)
```

```python
group2['response_rate']=group2['Yes']*100
```

```python
group2.describe().T
```

```python
bins = [-1, 20000,40000,60000,80000,100000000000000000000000000000000]
labels = ["[0:20000]","]20000:40000]","]40000:60000]","]60000:80000]","]80000: ]"]
group2['income_class'] = pd.cut(group2['income'], bins=bins, labels=labels)
```

```python
group2
```

```python
sns.barplot(x= group2['income_class'], y= group2['response_rate'] , data=group2)
plt.xlabel('Income')
plt.ylabel('Response rate')
plt.title('Response rate by income')
plt.show()
```
