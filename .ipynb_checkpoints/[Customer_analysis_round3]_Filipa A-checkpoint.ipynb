```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
marketing_data = pd.read_csv(r"C:\Users\pedro\Desktop\ironhack\LAB_costumer_analysis_round3\lab-customer-analysis-round-3\files_for_lab\csv_files\marketing_customer_analysis.csv")
marketing_data
```

```python
#standerdize columns.
cols= [] 

for i in range(len(marketing_data.columns)):
    cols.append(marketing_data.columns[i].lower().replace(' ','_'))

marketing_data.columns=cols

marketing_data

```

```python
#show columns, means knowing how many columns and how many rows this df has
marketing_data.shape
marketing_data
```

```python
#describe the data, present all the values for the numericla variables of this df  
marketing_data.describe()
```

```python
#plot a graph, it could be either a barplot of a histplot

plt.hist(marketing_data['response'], color='purple', align='mid')

plt.xlabel('responses')
plt.ylabel('Frequency')
plt.title('Response counting')

plt.show()
```

```python
#grouping the sales channel by response and count the response values putting them in a new DF caled response_sales
response_sales = marketing_data.groupby('sales_channel')['response'].value_counts(normalize=True).unstack().fillna(0)
response_sales
```

```python
#determining the percentage of the counted response answers yes and saving it in a new column in the new DF response_sales
response_sales['response_rate'] = response_sales['Yes']*100
```

```python
#resetting the index so the response is the new index

response_sales.reset_index(inplace=True)
response_sales
```

```python
#using bar plot to compare the tow variables. 
sns.barplot(x=response_sales['sales_channel'], y=response_sales['response_rate'])

plt.title('Boxpot of Sales Channel vs Response Rate')
plt.xlabel('Sales channel')
plt.ylabel('Response Rate %')
plt.show()
```

```python
#creating a new column to change the yes by number 1 
marketing_data['changed_response'] = np.where(marketing_data['response'] =='Yes', 1, 0)
marketing_data
```

```python
# creaing bins to help divide the data 
marketing_data['total_claim_bins_2']= np.where(marketing_data['total_claim_amount']<=200, '0-200', np.where(marketing_data['total_claim_amount']<=400, '200-400', np.where(marketing_data['total_claim_amount']<=600, '400-600', np.where(marketing_data['total_claim_amount']<=800, '600-800','>800'))))
marketing_data
```

```python
#group the bins by reponse, already changed for 0 and 1s and create a new df

data_total_claim = marketing_data.groupby(marketing_data['total_claim_bins_2'])['changed_response'].value_counts(normalize=True).unstack().fillna(0)

data_total_claim
```

```python
#Calculate percentage of response_Rate 
data_total_claim['response_rate'] = data_total_claim[1]*100
data_total_claim
```

```python
# Reset the index of the table
data_total_claim.reset_index(inplace=True)
data_total_claim
```

```python
#using a barplot to compare the 2 variables 
sns.barplot(x=data_total_claim['total_claim_bins_2'], y = data_total_claim['response_rate'], estimator='mean')

plt.title('Boxpot of Total Claim Amount vs Response Rate')
plt.xlabel('Total Claim amount')
plt.ylabel('Response Rate (%)')
plt.show()


```

```python
# to know how to set the bins 
marketing_data.describe()['income']
```

```python
marketing_data['income_bins']= np.where(marketing_data['income']<=1000, '0-1000', np.where(marketing_data['income']<=2000, '1000-2000', np.where(marketing_data['income']<=3000, '2000-3000', np.where(marketing_data['income']<=4000, '3000-4000',np.where(marketing_data['income']<=5000, '4000-5000', np.where(marketing_data['income']<=6000, '5000-6000', np.where(marketing_data['income']<=7000, '6000-7000', np.where(marketing_data['income']<=8000, '7000-8000', '>8000'))))))))
marketing_data
```

```python
data_income = marketing_data.groupby(marketing_data['income_bins'])['changed_response'].value_counts(normalize=True).unstack().fillna(0)

data_income
```

```python
data_income['response_rate'] = data_income[1]*100
data_income
```

```python
data_income.reset_index(inplace=True)
data_income
```

```python
sns.barplot(x=data_income['income_bins'], y=data_income['response_rate'])

plt.title('Boxplot of Income vs Response_rate')
plt.xlabel('Income')
plt.ylabel('Response Rate (%)')
plt.show()
```
