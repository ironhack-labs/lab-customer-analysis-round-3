```python
df_copy = pd.DataFrame({
    'Months Since Policy Inception': [1, 2, 3, 4],
    'Response': [100, 150, 120, 200]
})


df_copy['Response'] = pd.to_numeric(df_copy['Response'])


plt.bar(df_copy['Months Since Policy Inception'], df_copy['Response'], width=0.6
       )
plt.ylim(0, max(df_copy['Response']) + 1)

plt.xlabel('Months Since Policy Inception')
plt.ylabel('Total Response')
plt.title('Total Response per Month')

plt.show()
```

```python
df_copy = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales Channel': [100, 150, 120, 200]
})

plt.bar(df_copy['Month'], df_copy['Sales Channel'], color='red')

plt.xlabel('Month')
plt.ylabel('Sales Channel')
plt.title('Total Sales')

plt.show()
```

```python
df_copy = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Total Claim Amount': [100, 150, 120, 200]
})

plt.bar(df_copy['Month'], df_copy['Total Claim Amount'], color='green')

plt.xlabel('Month')
plt.ylabel('Total Claim Amount')
plt.title('Total Sales')

plt.show()
```

```python
data = {'Income': [30000, 50000, 70000, 30000, 50000, 70000],
        'Response': [1, 0, 1, 1, 0, 0]}
df_copy = pd.DataFrame(data)

# Calcular la tasa de respuesta por nivel de ingresos
response_rate_by_income = df_copy.groupby('Income')['Response'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Income', y='Response', data=response_rate_by_income)
plt.xlabel('Income')
plt.ylabel('Response Rate')
plt.title('Response Rate by Income')
plt.show()
```
