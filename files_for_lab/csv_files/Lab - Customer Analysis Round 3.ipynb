```python
# Show DataFrame info.

import pandas as pd

datalab2 = pd.read_csv('datalab2.csv')
pd.DataFrame(datalab2)
```

```python
# Describe DataFrame.

pd.DataFrame(datalab2).describe()
```

```python
# Show a plot of the total number of responses.

import seaborn as sns
import matplotlib.pyplot as plt

responses_counts = datalab2['response'].value_counts()
responses_counts.plot(kind='bar')
plt.title('Breakdown of responses')
plt.xlabel('Response type')
plt.ylabel('Number of responses')
plt.show
```

```python
# Show a plot of the response rate by the sales channel.

sales_channel = []
response_rate = []

for x in datalab2['sales channel'].unique():
    sales_channel.append(x)
    response_rate.append(len(datalab2[(datalab2['response']=='Yes') & (datalab2['sales channel']==x)])
                         /datalab2['sales channel'].value_counts()[x]*100)

plt.bar(x=sales_channel,height=response_rate)
plt.title('Response rate by channel')
plt.xlabel('Channel type')
plt.ylabel('Response rate')
plt.show
```

```python
# Show a plot of the response rate by the total claim amount.

datalab2['total claim amount'].describe()
```

```python
datalab2['total claim amount bucket'] = pd.cut(datalab2['total claim amount'], bins=[0,500,1000,1500,2000,2500,3000])

claim_amount = []
response_rate = []

for x in datalab2['total claim amount bucket'].unique():
    claim_amount.append(x)
    response_rate.append(len(datalab2[(datalab2['response']=='Yes') & (datalab2['total claim amount bucket']==x)])
                         /datalab2['total claim amount bucket'].value_counts()[x]*100)

claim_table = pd.DataFrame(list(zip(claim_amount,response_rate)), columns=['claim Amount', 'response rate'])
claim_table.plot(kind='bar')
plt.title('Breakdown of responses')
plt.xlabel('Claim amount')
plt.ylabel('Response rate')
plt.show
```

```python
# Show a plot of the response rate by income.

datalab2['income'].describe()
```

```python
datalab2['income bucket'] = pd.cut(datalab2['income'], include_lowest=True, bins=[0,10000,20000,30000,40000,50000,60000,70000,80000,90000,100000])

income_amount = []
response_rate = []

for x in datalab2['income bucket'].unique():
    income_amount.append(x)
    response_rate.append(len(datalab2[(datalab2['response']=='Yes') & (datalab2['income bucket']==x)])
                         /datalab2['income bucket'].value_counts()[x]*100)

income_table = pd.DataFrame(list(zip(income_amount,response_rate)), columns=['income amount', 'response rate'])
income_table.plot(kind='bar')
plt.title('Breakdown of responses')
plt.xlabel('Claim amount')
plt.ylabel('Response rate')
plt.show
```
