```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
mca2 = pd.read_csv("/Users/jxhn/Desktop/IRONHACK/Week_1/Day_4/Morning/Jahn-lab-dw-data-structuring-and-combining/mca.csv")
mca2
```

```python
mca2 = mca2.drop('Unnamed: 0', axis=1)
```

```python
mca2 = mca2.drop('unnamed:_0', axis=1)
```

```python
mca2.info()
```

```python
mca2.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
response_counts = mca2['response'].value_counts()

response_counts.plot(kind='bar')

plt.title('Total Number of Responses')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
mca2.groupby('sales_channel')['response'].count()
```

```python
response_count = mca2.groupby(['sales_channel', 'response']).size().unstack()


response_count.plot(kind='bar', stacked=True)


plt.title('Response Counts by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Count')

plt.show()
```

```python
mca2['response'] = pd.Categorical(mca2['response'])


plt.figure(figsize=(10, 6))


sns.histplot(data=mca2, x='total_claim_amount', hue='response', multiple='stack', binwidth=1000, edgecolor='black')


plt.title('Distribution of Total Claim Amount by Response')
plt.xlabel('Total Claim Amount')
plt.ylabel('Count')


plt.show()
```

```python
plt.figure(figsize=(10, 6))


sns.boxplot(data=mca2, x='response', y='income')


plt.title('Response Rate by Income')
plt.xlabel('Response')
plt.ylabel('Income')


plt.show()
```
