```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
# show dataframe
data.info()
```

```python
# describe dataframe 
data.describe()
```

```python
# Show a plot of the total number of responses
data['Response'].value_counts().plot(kind='bar')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
# Show a plot of the response rate by the sales channel
pd.crosstab(data['Response'], data['Sales Channel'], normalize='index').plot(kind='bar')
plt.title('Response Rate by Sales Channel')
plt.xlabel('Response')
plt.ylabel('Rate')
plt.show()

# Show a plot of the response rate by the total claim amount
data['Total Claim Amount Category'] = pd.cut(data['Total Claim Amount'], bins=10)
pd.crosstab(data['Response'], data['Total Claim Amount Category'], normalize='index').plot(kind='bar')
plt.title('Response Rate by Total Claim Amount')
plt.xlabel('Response')
plt.ylabel('Rate')
plt.show()

# Show a plot of the response rate by income
data['Income Category'] = pd.cut(data['Income'], bins=10)
pd.crosstab(data['Response'], data['Income Category'], normalize='index').plot(kind='bar')
plt.title('Response Rate by Income')
plt.xlabel('Response')
plt.ylabel('Rate')
plt.show()
```
