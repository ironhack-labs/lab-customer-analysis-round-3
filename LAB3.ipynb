# THIS TABLE SHOWS US INFORMATION REGARDING PEOPLE THAT SIGNED FOR AN INSURANCE POLICY FOR THEIR CARS. IT SHOWS PERSONAL INFORMATION AND INFORMATION RELATED TO THE VEHICLES OF THE CUSTOMERS.

# #SHOWING THE TOTAL NUMBER OF RESPONSES. DISCRIMINATING RESPONSES BY 'YES' or 'NO'. WE CAN EASILY CONCLUDE THAT THERE ARE MUCH MORE 'NO's' THAN 'YES's

```python
plt.hist(dados['response'])
plt.ylabel('Frequency')
plt.title('Responses Type')
plt.show()
```

# #SHOWING THE RECEIVED AND NOT RECEIVED RESPONSES BY THE DIFFERENT CHANNELS. WE CAN EASILY CONCLUDE THAT AGENTS WERE ABLE TO GET MORE RESPONSES, BRANCHES IN SECOND PLACE, CALL CENTERS IN THE THIRD PLACE AND IN THE LAST POSITION 'WEB'. THE AMOUNT OF NOT RECEIVED RESPONSES BY CHANNEL IS PROPORTIONAL TO THE RESPONSES RECEIVED. AS HIGHER THE RESPONSES RECEIVED, HIGHER THE NOT RECEIVED, ON EACH CHANNEL.

```python
dados1=dados.groupby(['sales_channel','response']).size()
dados1=dados1.unstack()
dados1.plot(kind='bar')
```

# WITH THIS GRAPH WE CAN CONCLUDE THAT THERE ARE NO MAJOR DIFFERENCES IN THE DISTRIBUTION OF ANSWERS BY INCOME. ON THE OTHER HAND, WE CAN CONCLUDE ONCE AGAIN THAT THERE ARE MUCH MORE 'NO'S' THAN 'YES'S'

```python

sns.catplot(data=dados, x='response', y='income')
plt.show()
```

# THIS GRAPH SHOWS THE DISTRIBUTION OF RESPONSES BY TOTAL_CLAIM_AMOUT. WE CAN EASILY UNDERSTAND THAT VAST AMOUNT OF PEOPLE IN THIS TABLE ARE CLAIMING LOWER AMOUNTS. THE VAST MAJORITY OF THE TOTAL CLAIM AMOUNT IS LOWER THAN 1500. THERE ARE NOT RESPONSES RECEIVED ABOVE 1500 AND THERE NOT MANY MISSING RESPONSES ON CASES WITH A CLAIM AMOUNT ABOVE 1500.

```python

sns.catplot(data=dados, x='response', y='total_claim_amount')
plt.show()
```
